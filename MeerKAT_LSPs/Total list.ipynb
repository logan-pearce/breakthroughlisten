```python
%matplotlib notebook
fig = plt.figure(figsize=(10,6))
#ax=fig.add_subplot(111)
ax = fig.add_axes([0.1, 0.1, 0.7, 0.85])
hb = ax.hexbin(dataframe['ra'], dataframe['decl'],gridsize=500,cmap='gist_gray',bins='log',mincnt=1,label=None)

# MeerTime 1000:
k=pd.read_csv('MeerTime_1000PTA/meertime_1000PTA_1000brightest_targets.csv')
ax.scatter(k['RAJD'],k['DECJD'],alpha=0.8,color='blue',s=10,marker='.',label='MeerTime\n1000PTA')
# MeerTime MSP:
k=pd.read_csv('MeerTime_MSP/meertime_msp_targets.csv')
ax.scatter(k['ra_deg'][np.where(k['dec_deg'].values<45.)[0]],k['dec_deg'][np.where(k['dec_deg'].values<45.)[0]]\
           ,alpha=0.8,color='purple',s=10,marker='.',label='MeerTime\nMSP')
# MeerTime Binary:
k=pd.read_csv('MeerTime_binaries/MeerTime_binaries_atnfcatalog_notgc.csv')
ax.scatter(k['RAJD'],k['DECJD'],alpha=0.8,color='dodgerblue',s=10,marker='.',label='MeerTime\nBinaries')
# MeerTime GC:
k=pd.read_csv('MeerTime_GC/MeerTime_GC_targets.csv')
ax.scatter(k['ra_deg'],k['dec_deg'],alpha=0.8,color='lightseagreen',s=10,marker='.',\
           label='MeerTime\nGlobular\nClusters')
# Mightee:
ax.fill(M_RA_1, M_Dec_1, "red", alpha=0.5,label="MIGHTEE",zorder=10)
ax.fill(M_RA_2, M_Dec_2, "red", alpha=0.5,zorder=10)
ax.scatter(150.119,2.2,alpha=1.0,color='red',s=10,marker='.')
# Fornax:
RA,Dec = 50.67, -37.2
box_x = [RA-5,RA+3]
box_y = [Dec-3.,Dec+2]
ax.fill([box_x[0],box_x[0],box_x[1],box_x[1],box_x[0]],[box_y[0],box_y[1],box_y[1],box_y[0],box_y[0]],\
       "green", alpha=0.5,label="Fornax",zorder=10)
# Mhongoose:
k=pd.read_csv('Mhongoose/mhongoose_webtable.csv')
ax.scatter(k['RA_deg'],k['Dec_deg'],alpha=0.8,color='yellow',s=10,marker='.',label='Mhongoose')
# Laduma:
ax.scatter(53.12, -28.13,alpha=1.0,color='magenta',s=10,marker='.',label='LADUMA')
# Trapum GC:
k=pd.read_csv('Trapum_GC/Trapum_GC_targets.csv')
ax.scatter(k['ra_deg'],k['dec_deg'],alpha=0.8,color='darkorange',s=10,marker='.',label='Trapum\nGlobular\nClusters')
# Trapum Fermi:
k=pd.read_csv('Trapum_Fermi/Trapum_fermi_targets.csv')
ax.scatter(k['RA'],k['Dec'],alpha=0.8,color='salmon',s=10,marker='.',label='Trapum\nFermi\nSources')
# Trapum Nearby Galaxies:
ax.scatter(T_RA,T_Dec,alpha=0.8,color='brown',s=40,marker='o',label='Trapum\nNearby\nGalaxies')

cb = fig.colorbar(hb, ax=ax,orientation="horizontal", pad=0.1)
cb.set_label('log10(count)')
ax.grid(True)
ax.legend(bbox_to_anchor=(1.01, 0.39,0.2,0.6), loc=2,
           ncol=1, mode="expand", borderaxespad=0.,fontsize=10)
ax.set_xlabel('RA (deg)')
ax.set_ylabel('DEC (deg)')
#plt.tight_layout()
plt.show()
```
