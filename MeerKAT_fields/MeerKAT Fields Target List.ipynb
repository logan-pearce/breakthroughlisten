```python
import time
start = time.time()
d=np.zeros(len(color))
print 'Computing distances'
count=0
L=1350
for i in range(len(d)):
    coeff = np.array([(1./L),(-2),((f['parallax'].values[i]/1000.)/((f['parallax_error'].values[i]/1000.))**2),\
                      -(1./((f['parallax_error'].values[i]/1000.)**2))])
    g = np.roots(coeff)
    # Find the number of real roots:
    reals = np.isreal(g)
    realsum = np.sum(reals)
    # If there is one real root, that root is the  mode:
    if realsum == 1:
        gd = np.real(g[np.where(reals)[0]])
    # If all roots are real:
    elif realsum == 3:
        if f['parallax'].values[i] >= 0:
            # Take the smallest root:
            gd = np.min(g)
        elif f['parallax'].values[i] < 0:
            # Take the positive root (there should be only one):
            gd = g[np.where(g>0)[0]]
    d[i] = gd
    
    count = count+1
    mod=count%10000
    if mod==0:
        print 'Finished ',count,' sources'
print 'Finished distances for ',d.shape[0]

end = time.time()
print(end - start)

print 'Computing FWHM'

count=0
rmax = 1e6
fwhm_lo,fwhm_hi = np.zeros(len(d)),np.zeros(len(d))
sigma = np.zeros(len(d))
from scipy.optimize import brentq
for i in range(len(d)):
    rmode = d[i]
    M = (rmode**2*np.exp(-rmode/L)/(f['parallax_error'].values[i]/1000.))*\
        np.exp((-1./(2*(f['parallax_error'].values[i]/1000.)**2))*(f['parallax'].values[i]/1000.-(1./rmode))**2)
    lo = brentq(lambda x: 2*np.log(x)-(x/L)-((((f['parallax'].values[i]/1000.)-(1./x))**2)/(2*(f['parallax_error'].values[i]/1000.)**2)) \
               +np.log(2)-np.log(M)-np.log(f['parallax_error'].values[i]/1000.), 0.001, rmode)
    hi = brentq(lambda x: 2*np.log(x)-(x/L)-((((f['parallax'].values[i]/1000.)-(1./x))**2)/(2*(f['parallax_error'].values[i]/1000.)**2)) \
               +np.log(2)-np.log(M)-np.log(f['parallax_error'].values[i]/1000.), rmode, rmax)
    fwhm_lo[i],fwhm_hi[i] = lo,hi
    
    sigma[i] = (hi-lo)/2.355
    
    count = count+1
    mod=count%10000
    if mod==0:
        print 'Finished ',count,' sources'

print 'Finished FWHM for ',fwhm_lo.shape[0]
end = time.time()
print(end - start)
```

```python
db = MySQLdb.connect(host='104.154.94.28',db='loganp',\
                     read_default_file="~/.my.cnf",\
                     autocommit=True,\
                     local_infile = 1)
c=db.cursor()
cursor=db.cursor()
```

```python
# Create new columns on the table to be edited:
string = "ALTER TABLE meerkat_gaiadr2_mightee \
            ADD `dist.c` FLOAT NOT NULL,\
            ADD `fwhm_lo.c` FLOAT NOT NULL,\
            ADD `fwhm_hi.c` FLOAT NOT NULL,\
            ADD `sd.c` FLOAT NOT NULL"
c.execute(string)
```

```python
L=1350
rmax = 1e6
c.execute("""SELECT source_id,parallax,parallax_error,bp_rp,phot_g_mean_mag,ra,decl \
                FROM meerkat_gaiadr2_laduma\
                WHERE 1 """)
numrows = c.rowcount
```

Pull out one row at a time from the query results.  Perform the math on just that row, then update the corresponding cell in the new columns of the original table that match with the same source id from the original table.

NOTE - It is exceedingly faster if you index the source_id column on the original table first.  This lets MySQL search for the matching row much more quickly.  In phpmyadmin, click on "Structure", then click the little lightning bolt logo called "Index" in the source_id row.  Do this before executing the following cell.  This is much more important for the much larger tables.

```python
for x in xrange(0,numrows):
    row = c.fetchone()
    # Compute distance:
    coeff = np.array([(1./L),(-2),((row[1]/1000.)/((row[2]/1000.))**2),\
                      -(1./((row[2]/1000.)**2))])
    g = np.roots(coeff)
    # Find the number of real roots:
    reals = np.isreal(g)
    realsum = np.sum(reals)
    # If there is one real root, that root is the  mode:
    if realsum == 1:
        d = np.real(g[np.where(reals)[0]])
    # If all roots are real:
    elif realsum == 3:
        if row[1] >= 0:
            # Take the smallest root:
            d = np.min(g)
        elif row[1] < 0:
            # Take the positive root (there should be only one):
            d = g[np.where(g>0)[0]]
    # Compute FWHM: 
    rmode = d
    M = (rmode**2*np.exp(-rmode/L)/(row[2]/1000.))*\
        np.exp((-1./(2*(row[2]/1000.)**2))*(row[1]/1000.-(1./rmode))**2)
    lo = brentq(lambda x: 2*np.log(x)-(x/L)-((((row[1]/1000.)-(1./x))**2)/(2*(row[2]/1000.)**2)) \
               +np.log(2)-np.log(M)-np.log(row[2]/1000.), 0.001, rmode)
    hi = brentq(lambda x: 2*np.log(x)-(x/L)-((((row[1]/1000.)-(1./x))**2)/(2*(row[2]/1000.)**2)) \
               +np.log(2)-np.log(M)-np.log(row[2]/1000.), rmode, rmax)
    # Compute std dev:
    sigma = (hi-lo)/2.355
    
    # Update the original table with the new values, matching by source id:
    string = """UPDATE `meerkat_gaiadr2_laduma` SET `dist.c`="""+str(d[0])+""" \
                    WHERE `source_id` ="""+str(row[0])
    cursor.execute(string)
    string = """UPDATE `meerkat_gaiadr2_laduma` SET `fwhm_lo.c`="""+str(lo)+""" \
                    WHERE `source_id` ="""+str(row[0])
    cursor.execute(string)
    string = """UPDATE `meerkat_gaiadr2_laduma` SET `fwhm_hi.c`="""+str(hi)+""" \
                    WHERE `source_id` ="""+str(row[0])
    cursor.execute(string)
    string = """UPDATE `meerkat_gaiadr2_laduma` SET `sd.c`="""+str(sigma)+""" \
                    WHERE `source_id` ="""+str(row[0])
    cursor.execute(string)
```
